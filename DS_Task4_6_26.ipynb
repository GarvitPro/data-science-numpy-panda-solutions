```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', hue='day', data=df)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(x='size', y='total_bill', data=df)
plt.title("Group Size vs Total Bill")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Boxplot of Bills by Day")
plt.show()
```

```python
sns.pairplot(df, hue='sex')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['total_bill'], df['tip'], df['size'])

ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
ax.set_zlabel('Size')

plt.title("3D Scatter Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(df['total_bill'], df['tip'], df['size'])

plt.title("3D Line Plot")
plt.show()
```

```python
from matplotlib import cm

X = np.linspace(df['total_bill'].min(), df['total_bill'].max(), 30)
Y = np.linspace(df['tip'].min(), df['tip'].max(), 30)

X, Y = np.meshgrid(X, Y)

Z = np.sin(X/10) + np.cos(Y)  # artificial surface

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

plt.title("3D Surface Plot")
plt.show()
```
